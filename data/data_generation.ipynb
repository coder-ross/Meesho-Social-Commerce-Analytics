```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

print("🚀 Generating Realistic Meesho Dataset...")
print("=" * 60)

# ============================================================================
# 1. SUPPLIERS TABLE
# ============================================================================
print("\n📦 Generating Suppliers...")

supplier_cities = ['Delhi', 'Mumbai', 'Surat', 'Jaipur', 'Bangalore', 'Ahmedabad', 
                   'Ludhiana', 'Tirupur', 'Indore', 'Chennai', 'Kolkata', 'Panipat']

city_to_state = {
    'Delhi': 'Delhi', 'Mumbai': 'Maharashtra', 'Surat': 'Gujarat',
    'Jaipur': 'Rajasthan', 'Bangalore': 'Karnataka', 'Ahmedabad': 'Gujarat',
    'Ludhiana': 'Punjab', 'Tirupur': 'Tamil Nadu', 'Indore': 'Madhya Pradesh',
    'Chennai': 'Tamil Nadu', 'Kolkata': 'West Bengal', 'Panipat': 'Haryana'
}

suppliers_list = []
for i in range(1, 151):
    city = np.random.choice(supplier_cities)
    suppliers_list.append({
        'supplier_id': i,
        'supplier_name': f'Supplier_{i:03d}',
        'city': city,
        'state': city_to_state[city],
        'registration_date': datetime(2021, 1, 1) + timedelta(days=i*3),
        'supplier_tier': np.random.choice(['Gold', 'Silver', 'Bronze'], p=[0.2, 0.5, 0.3]),
        'avg_rating': round(np.random.uniform(3.5, 4.9), 2)
    })

suppliers = pd.DataFrame(suppliers_list)
print(f"✅ Created {len(suppliers)} suppliers")

# ============================================================================
# 2. RESELLERS TABLE
# ============================================================================
print("\n👥 Generating Resellers...")

tier1_cities = ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai', 'Kolkata', 'Pune']
tier2_cities = ['Jaipur', 'Lucknow', 'Indore', 'Bhopal', 'Patna', 'Nagpur', 'Vadodara', 
                'Coimbatore', 'Ludhiana', 'Agra', 'Kochi', 'Visakhapatnam']
tier3_cities = ['Meerut', 'Rajkot', 'Varanasi', 'Amritsar', 'Allahabad', 'Ranchi',
                'Gwalior', 'Jabalpur', 'Vijayawada', 'Jodhpur', 'Raipur', 'Mysore']

all_cities_map = {
    'Mumbai': ('Maharashtra', 'Tier 1'), 'Delhi': ('Delhi', 'Tier 1'),
    'Bangalore': ('Karnataka', 'Tier 1'), 'Hyderabad': ('Telangana', 'Tier 1'),
    'Chennai': ('Tamil Nadu', 'Tier 1'), 'Kolkata': ('West Bengal', 'Tier 1'),
    'Pune': ('Maharashtra', 'Tier 1'),
    'Jaipur': ('Rajasthan', 'Tier 2'), 'Lucknow': ('Uttar Pradesh', 'Tier 2'),
    'Indore': ('Madhya Pradesh', 'Tier 2'), 'Bhopal': ('Madhya Pradesh', 'Tier 2'),
    'Patna': ('Bihar', 'Tier 2'), 'Nagpur': ('Maharashtra', 'Tier 2'),
    'Vadodara': ('Gujarat', 'Tier 2'), 'Coimbatore': ('Tamil Nadu', 'Tier 2'),
    'Ludhiana': ('Punjab', 'Tier 2'), 'Agra': ('Uttar Pradesh', 'Tier 2'),
    'Kochi': ('Kerala', 'Tier 2'), 'Visakhapatnam': ('Andhra Pradesh', 'Tier 2'),
    'Meerut': ('Uttar Pradesh', 'Tier 3'), 'Rajkot': ('Gujarat', 'Tier 3'),
    'Varanasi': ('Uttar Pradesh', 'Tier 3'), 'Amritsar': ('Punjab', 'Tier 3'),
    'Allahabad': ('Uttar Pradesh', 'Tier 3'), 'Ranchi': ('Jharkhand', 'Tier 3'),
    'Gwalior': ('Madhya Pradesh', 'Tier 3'), 'Jabalpur': ('Madhya Pradesh', 'Tier 3'),
    'Vijayawada': ('Andhra Pradesh', 'Tier 3'), 'Jodhpur': ('Rajasthan', 'Tier 3'),
    'Raipur': ('Chhattisgarh', 'Tier 3'), 'Mysore': ('Karnataka', 'Tier 3')
}

resellers_list = []
for i in range(1, 1001):
    tier_choice = random.choices(['Tier 1', 'Tier 2', 'Tier 3'], weights=[0.15, 0.45, 0.40])[0]
    
    if tier_choice == 'Tier 1':
        city = random.choice(tier1_cities)
    elif tier_choice == 'Tier 2':
        city = random.choice(tier2_cities)
    else:
        city = random.choice(tier3_cities)
    
    state, tier = all_cities_map[city]
    
    resellers_list.append({
        'reseller_id': i,
        'reseller_name': f'Reseller_{i:04d}',
        'phone': f'+91{random.randint(7000000000, 9999999999)}',
        'city': city,
        'state': state,
        'tier': tier,
        'join_date': datetime(2022, 6, 1) + timedelta(hours=i*12),
        'reseller_status': np.random.choice(['Active', 'Inactive', 'Churned'], p=[0.55, 0.30, 0.15])
    })

resellers = pd.DataFrame(resellers_list)
print(f"✅ Created {len(resellers)} resellers")
print(f"   - Tier 1: {len(resellers[resellers['tier']=='Tier 1'])}")
print(f"   - Tier 2: {len(resellers[resellers['tier']=='Tier 2'])}")
print(f"   - Tier 3: {len(resellers[resellers['tier']=='Tier 3'])}")

# ============================================================================
# 3. PRODUCTS TABLE
# ============================================================================
print("\n🛍️ Generating Products...")

categories = {
    'Fashion': ['Womens Ethnic', 'Mens Casual', 'Kids Wear', 'Sarees', 'Kurtis', 'T-Shirts'],
    'Home & Kitchen': ['Kitchen Tools', 'Home Decor', 'Bedding', 'Storage'],
    'Beauty & Personal Care': ['Makeup', 'Skincare', 'Haircare', 'Fragrances'],
    'Electronics': ['Mobile Accessories', 'Gadgets', 'Electronics']
}

product_list = []
product_id = 1

for category, subcategories in categories.items():
    if category == 'Fashion':
        num_products = 1250
    elif category == 'Home & Kitchen':
        num_products = 625
    elif category == 'Beauty & Personal Care':
        num_products = 375
    else:
        num_products = 250
    
    for _ in range(num_products):
        subcategory = random.choice(subcategories)
        cost_price = round(random.uniform(100, 2000), 2)
        mrp = round(cost_price * random.uniform(1.4, 2.5), 2)
        
        product_list.append({
            'product_id': product_id,
            'supplier_id': random.randint(1, 150),
            'product_name': f'{subcategory}_{product_id:04d}',
            'category': category,
            'subcategory': subcategory,
            'cost_price': cost_price,
            'mrp': mrp,
            'commission_percent': round(random.uniform(5, 25), 2)
        })
        product_id += 1

products = pd.DataFrame(product_list)
print(f"✅ Created {len(products)} products")

# ============================================================================
# 4. ORDERS TABLE
# ============================================================================
print("\n📊 Generating Orders...")

start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 6, 30)

date_weights = []
current = start_date
while current <= end_date:
    month = current.month
    if month in [10, 11]:
        weight = 2.0
    elif month in [2, 8, 3]:
        weight = 1.5
    else:
        weight = 1.0
    date_weights.append((current, weight))
    current += timedelta(days=1)

order_list = []
active_resellers = resellers[resellers['reseller_status'] == 'Active']['reseller_id'].tolist()
other_resellers = resellers[resellers['reseller_status'] != 'Active']['reseller_id'].tolist()

for order_id in range(1, 25001):
    order_date = random.choices([d[0] for d in date_weights], 
                                weights=[d[1] for d in date_weights])[0]
    
    reseller_id = random.choices(
        active_resellers + other_resellers,
        weights=[3]*len(active_resellers) + [1]*len(other_resellers)
    )[0]
    
    reseller_data = resellers[resellers['reseller_id'] == reseller_id].iloc[0]
    delivery_date = order_date + timedelta(days=random.randint(3, 10))
    
    order_list.append({
        'order_id': order_id,
        'reseller_id': reseller_id,
        'customer_name': f'Customer_{order_id:05d}',
        'order_date': order_date,
        'delivery_date': delivery_date if random.random() > 0.05 else None,
        'city': reseller_data['city'],
        'state': reseller_data['state'],
        'order_status': random.choices(['Delivered', 'Returned', 'Cancelled'], 
                                      weights=[0.85, 0.10, 0.05])[0],
        'total_amount': 0,
        'payment_method': random.choice(['COD', 'UPI', 'Card', 'Wallet'])
    })

orders = pd.DataFrame(order_list)
orders['total_amount'] = orders['total_amount'].astype(float)

print(f"✅ Created {len(orders)} orders")

# ============================================================================
# 5. ORDER ITEMS TABLE
# ============================================================================
print("\n📦 Generating Order Items...")

order_items_list = []
order_item_id = 1

for _, order in orders.iterrows():
    num_items = random.choices([1, 2, 3, 4, 5], weights=[0.45, 0.30, 0.15, 0.07, 0.03])[0]
    order_total = 0
    
    for _ in range(num_items):
        product = products.sample(1).iloc[0]
        quantity = random.choices([1, 2, 3], weights=[0.75, 0.20, 0.05])[0]
        discount = random.uniform(5, 40)
        price_sold = round(product['mrp'] * (1 - discount/100), 2)
        
        order_items_list.append({
            'order_item_id': order_item_id,
            'order_id': order['order_id'],
            'product_id': product['product_id'],
            'quantity': quantity,
            'price_sold': price_sold,
            'discount_percent': round(discount, 2)
        })
        
        order_total += price_sold * quantity
        order_item_id += 1
    
    orders.loc[orders['order_id'] == order['order_id'], 'total_amount'] = round(order_total, 2)

order_items = pd.DataFrame(order_items_list)
print(f"✅ Created {len(order_items)} order items")

# ============================================================================
# 6. RESELLER ACTIVITY TABLE
# ============================================================================
print("\n📱 Generating Reseller Activity...")

activity_list = []
activity_id = 1

for _, reseller in resellers.iterrows():
    current_date = reseller['join_date']
    end = datetime(2024, 6, 30)
    
    while current_date <= end:
        if reseller['reseller_status'] == 'Active':
            products_shared = random.randint(20, 150)
            catalog_views = random.randint(50, 500)
            orders_generated = random.randint(2, 20)
        elif reseller['reseller_status'] == 'Inactive':
            products_shared = random.randint(0, 30)
            catalog_views = random.randint(5, 100)
            orders_generated = random.randint(0, 5)
        else:
            products_shared = 0
            catalog_views = random.randint(0, 10)
            orders_generated = 0
        
        activity_list.append({
            'activity_id': activity_id,
            'reseller_id': reseller['reseller_id'],
            'activity_date': current_date,
            'products_shared': products_shared,
            'catalog_views': catalog_views,
            'orders_generated': orders_generated
        })
        
        activity_id += 1
        current_date += timedelta(days=30)

reseller_activity = pd.DataFrame(activity_list)
print(f"✅ Created {len(reseller_activity)} activity records")

# ============================================================================
# 7. RETURNS TABLE
# ============================================================================
print("\n↩️ Generating Returns...")

returned_orders = orders[orders['order_status'] == 'Returned']
return_reasons = ['Size Issue', 'Quality Issue', 'Wrong Product', 'Damaged in Transit',
                  'Not as Described', 'Changed Mind', 'Defective Product']

returns_list = []
for idx, (_, order) in enumerate(returned_orders.iterrows(), 1):
    return_date = None
    if pd.notna(order['delivery_date']):
        return_date = order['delivery_date'] + timedelta(days=random.randint(1, 7))
    
    returns_list.append({
        'return_id': idx,
        'order_id': order['order_id'],
        'return_date': return_date,
        'return_reason': random.choice(return_reasons),
        'refund_amount': order['total_amount']
    })

returns = pd.DataFrame(returns_list)
print(f"✅ Created {len(returns)} returns")

# ============================================================================
# SAVE ALL FILES
# ============================================================================
print("\n💾 Saving CSV Files...")

suppliers.to_csv('suppliers.csv', index=False)
resellers.to_csv('resellers.csv', index=False)
products.to_csv('products.csv', index=False)
orders.to_csv('orders.csv', index=False)
order_items.to_csv('order_items.csv', index=False)
reseller_activity.to_csv('reseller_activity.csv', index=False)
returns.to_csv('returns.csv', index=False)

print("\n" + "=" * 60)
print("✨ DATASET GENERATION COMPLETE! ✨")
print("=" * 60)
print("\nFiles Created:")
print("  1. suppliers.csv")
print("  2. resellers.csv")
print("  3. products.csv")
print("  4. orders.csv")
print("  5. order_items.csv")
print("  6. reseller_activity.csv")
print("  7. returns.csv")
print(f"\n📊 Total GMV: ₹{orders[orders['order_status']=='Delivered']['total_amount'].sum():,.2f}")
print(f"Average Order Value: ₹{orders['total_amount'].mean():,.2f}")
print("🚀 Ready for analysis!")
```
